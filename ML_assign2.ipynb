### (b) Create a new Data Frame with the selected attributes only.

```python
removed_features = [
    "Title",
    "FirstName",
    "MiddleName",
    "LastName",
    "Suffix",
    "AddressLine1",
    "AddressLine2",
    "City",
    "StateProvinceName",
    "PostalCode",
    "PhoneNumber",
    "LastUpdated",
]
```

```python
from datetime import date

def calculate_float_age(born):
    today = pd.to_datetime(date.today())
    days_lived = (today - born).days
    return days_lived / 365.25   # 365.25 accounts for leap years
```

```python
df_new = df.drop(removed_features, axis=1)

df_new["BirthDate"] = pd.to_datetime(df_new["BirthDate"])
df_new["BirthDate"] = df_new["BirthDate"].apply(calculate_float_age)
df_new.rename(columns={"BirthDate": "Age"}, inplace=True)

df_new.rename(columns={"CountryRegionName": "Country"}, inplace=True)

df_new["YearlyIncome"] = df_new["YearlyIncome"].astype(float)

df_new.head()
```

### (d) Standardization/Normalization

```python
from sklearn.preprocessing import StandardScaler

df_normalised = df_new.copy(deep=True)

transformed_cols = [col for col in numeric_cols if col not in target_cols]

sc = StandardScaler()
df_normalised[transformed_cols] = sc.fit_transform(df_new[transformed_cols])

df_normalised.head()
```

### (e) Binarization (One Hot Encoding)

```python
# Columns to be binarised = Non-numeric categorical columns

binarising_cols = [
    col for col in categorical_cols
    if col not in target_cols
    and col not in df_normalised.select_dtypes(np.number)
]

df_normalised[binarising_cols].nunique()
```

```python
unique_vals = []
for col in binarising_cols:
    unique_vals.append({"Column": col, "Unique Vals": df_normalised[col].unique().tolist()})

pd.DataFrame(unique_vals).set_index("Column")
```

```python
# Handling binary features

df_binarized = df_normalised.copy(deep=True)

df_binarized["MaritalStatus"] = df_normalised["MaritalStatus"].map({"M": 1, "S": 0})
df_binarized.rename({"MaritalStatus": "IsMarried"}, axis=1, inplace=True)

df_binarized["Gender"] = df_normalised["Gender"].map({"M": 1, "F": 0})
df_binarized.rename({"Gender": "IsMale"}, axis=1, inplace=True)

df_binarized.head(3)
```

```python
# Handling Education as ordinal data

df_order = df_binarized.copy(deep=True)

df_order["Education"] = df_order["Education"].map({
    "Partial High School": 0,
    "High School": 1,
    "Partial College": 2,
    "Bachelors": 3,
    "Graduate Degree": 4,
})
df_order.rename({"Education": "EducationLevel"}, axis=1, inplace=True)

df_order.head(3)
```

```python
# Applying dummy encoding

encode_cols = ["Country", "Occupation"]
```

```python
# METHOD 1: Using OneHotEncoder from SciKit-Learn

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse_output=False) # sparse_output=False returns a dense NumPy array!
encoded = encoder.fit_transform(df_order[encode_cols])

encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(), index=df_order.index)
df_encoded = pd.concat([df_order.drop(columns=encode_cols), encoded_df], axis=1)

df_encoded.head()
```

```python
# METHOD 2: Using Panda's get_dummies()

df_encoded = pd.get_dummies(data=df_order, columns=encode_cols, dtype=int, drop_first=True)
df_encoded.head()
```

### (a) Calculate Similarity in Simple Matching, Jaccard Similarity, and Cosine Similarity between two following objects of your transformed input data.

```python
# Calculating similarity between 2 binary columns

sim_cols = ["IsMale", "BikeBuyer"]
```

```python
# Simple Matching Coefficient, SMC = 11 + 00 / 11 + 10 + 01 + 00

# Hamming loss = Sum(A != B) / n = 10 + 01 / 11 + 10 + 01 + 00
# => SMC = 1 - Hamming Loss

from sklearn.metrics import hamming_loss
SMC = 1 - hamming_loss(df_final[sim_cols[0]], df_final[sim_cols[1]])

print(f"Simple Matching Coefficient between {sim_cols[0]} and {sim_cols[1]} = {SMC}")
```

```python
# Jaccard Similarity = M11 / M11 + M10 + M01

from sklearn.metrics import jaccard_score
jaccard = jaccard_score(df_final[sim_cols[0]], df_final[sim_cols[1]])

print(f"Jaccard similarity between {sim_cols[0]} and {sim_cols[1]} = {jaccard}")
```

```python
# Cosine Similarity = A.B / ||A||.||B||

from sklearn.metrics.pairwise import cosine_similarity
cosine = cosine_similarity(df_final[sim_cols].T)[0][1]
# cosine_similarity expects row vectors by default, and returns matrix like similarities between rows 0 & 0, 0 & 1, 1 & 0, and 1 & 1 in each cell!

print(f"Cosine similarity between {sim_cols[0]} and {sim_cols[1]} = {cosine}")
```
