**Features REMOVED (14 attributes):**

### (b) Create a new Data Frame with the selected attributes only.

```python
removed_features = [
    "Title",
    "FirstName",
    "MiddleName",
    "LastName",
    "Suffix",
    "AddressLine1",
    "AddressLine2",
    "PostalCode",
    "PhoneNumber",
    "LastUpdated",
]
```

```python
from datetime import date
def calculate_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
```

```python
df_train_new = df_train.drop(removed_features, axis=1)

df_train_new["BirthDate"] = pd.to_datetime(df_train_new["BirthDate"])
df_train_new["BirthDate"] = df_train_new["BirthDate"].apply(calculate_age)
df_train_new.rename(columns={"BirthDate": "Age"}, inplace=True)

df_train_new["YearlyIncome"] = df_train_new["YearlyIncome"].astype(float)

df_train_new.head()
```

```python
df_test_classif_new = df_test_classif.drop(removed_features, axis=1)

df_test_classif_new["BirthDate"] = pd.to_datetime(df_test_classif_new["BirthDate"])
df_test_classif_new["BirthDate"] = df_test_classif_new["BirthDate"].apply(calculate_age)
df_test_classif_new.rename(columns={"BirthDate": "Age"}, inplace=True)

df_test_classif_new["YearlyIncome"] = df_test_classif_new["YearlyIncome"].astype(float)

df_test_classif_new.head()
```

```python
df_test_regress_new = df_test_regress.drop(removed_features, axis=1)

df_test_regress_new["BirthDate"] = pd.to_datetime(df_test_regress_new["BirthDate"])
df_test_regress_new["BirthDate"] = df_test_regress_new["BirthDate"].apply(calculate_age)
df_test_regress_new.rename(columns={"BirthDate": "Age"}, inplace=True)

df_test_regress_new["YearlyIncome"] = df_test_regress_new["YearlyIncome"].astype(float)

df_test_regress_new.head()
```
